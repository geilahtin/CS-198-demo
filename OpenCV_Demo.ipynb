To import the OpenCV library, use the command "import cv2"

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

Example 1: Read and display image. <br>
The command to read image is cv2.imread(path, flag).

```python
image = cv2.imread("bunny.jpg")   # Read the image
plt.imshow(image)
```

The reason why it is in that format, OpenCV follows the BGR format. The reason is that back then the format is popular among camera manufacturers and software providers. To convert to RGB format, run the command below.

Example 2: Convert to RGB image <br>
The command is cv2.cvtColor(src, code[, dst[, dstCn]])

```python
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_RGB)
```

Example 3: Drawing shapes into a picture

The parameters in drawing a rectangle is cv2.rectangle(image, start_point, end_point, color, thickness)

```python
rectangle_image= image_RGB.copy()

cv2.rectangle(rectangle_image,(25,50),(175,150),(0,0,255),3)

plt.imshow(rectangle_image)
```

The parameters used in drawing a line is cv2.line(image, start_point, end_point, color, thickness) 

```python
line_image= image_RGB.copy()

cv2.line(line_image,(25,50),(175,150),(0,0,255),3)

plt.imshow(line_image)
```

The parameters used in drawing a circle is cv2.circle(image, center_coordinates, radius, color, thickness)

```python
circle_image= image_RGB.copy()

cv2.circle(circle_image,(100,130), 90, (0,0,255), -1)

plt.imshow(circle_image)
```

Example 4: Erosion and Dilation

Erosion and Dilation are morphological operations. 

The parameters used for erosion and dilation are cv2.erode(original image, kernel, iterations) and cv2.dilate(original image, kernel, iterations)

```python
import numpy as np

img = cv2.imread('bunny.jpg', 0)
 
# Taking a matrix of size 5 as the kernel
kernel = np.ones((5, 5), np.uint8)
 
img_erosion = cv2.erode(img, kernel, iterations=1)
 
plt.imshow(img_erosion)
```

```python
import numpy as np

img = cv2.imread('bunny.jpg', 0)
 
# Taking a matrix of size 5 as the kernel
kernel = np.ones((5, 5), np.uint8)
 
img_dilation = cv2.dilate(img, kernel, iterations=1)
 
plt.imshow(img_dilation)
```

Example 5: Getting live video capture

```python
vid = cv2.VideoCapture(0)
  
while(True):
      
    # Capture the video frame by frame
    ret, frame = vid.read()
  
    # Display the resulting frame
    cv2.imshow('frame', frame)
      
    # The 'q' button is set as the quitting button desired button of your choice
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
  
# After the loop release the cap object
vid.release()

# Destroy all the windows
cv2.destroyAllWindows()
```

Example 6: Extracting images from video

```python
import os

cam = cv2.VideoCapture("MitchiriNeko March.mp4")
  
try:   
    # creating a folder named data
    if not os.path.exists('data'):
        os.makedirs('data')
  
except OSError:
    print ('Error: Creating directory of data')
  
currentframe = 0
  
while(True):
      
    # reading from frame
    ret,frame = cam.read()
  
    if ret:
        # if video is still left continue creating images
        name = './data/frame' + str(currentframe) + '.jpg'
        print ('Creating...' + name)
  
        # saving the extracted images
        cv2.imwrite(name, frame)
  
        # increasing counter so that it will show how many frames are created
        currentframe += 1
    else:
        break
  
cam.release()
cv2.destroyAllWindows()
```
